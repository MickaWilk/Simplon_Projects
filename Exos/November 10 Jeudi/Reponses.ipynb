```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("lancers.csv", index_col=0)
df.dtypes
```

```python
df = ((df.applymap(str)).applymap(lambda x: x.replace(",",".") if (type(x) == type("")) else x)).applymap(float)
df
```

```python
x = df.to_numpy().reshape(-1)
plt.hist(x)
```

```python
fig,ax = plt.subplots()
for i in df:
    ax.plot(i) 
```
